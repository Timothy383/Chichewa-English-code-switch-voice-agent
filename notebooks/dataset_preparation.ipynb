```python
# import necessary ribraries
import pandas as pd
import os
import whisper
from tqdm import tqdm  # shows a progress bar
```

```python
# Paths
audio_folder = "chichewa_voice_agent/audios"
training_csv = "chichewa_voice_agent/phrases/phrases.csv"
```

```python
# load phrases dataset
df = pd.read_csv(training_csv)
df.head()  # check first few rows
```

```python
# create full audio path
df["audio_path"] = df["audio_file"].apply(
    lambda x: os.path.join(audio_folder, x)
)
```

```python
# prepare dataset for Whisper
df_finetune = df[["audio_path", "phrase"]].copy()
```

```python
# rename column to expected name
df_finetune.rename(columns={"phrase": "text"}, inplace=True)
```

```python
# save dataset
df_finetune.to_csv("chichewa_whisper_finetune.csv", index=False)
```
